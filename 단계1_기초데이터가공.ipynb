기초 데이터 가공이란?
 - 메모리에 올려 놓고 사용하기 위한 전처리 작업을 뜻함
   매번 파일을 읽어 변수에 넣어두는 작업이 장시간 소요되므로, 이 작업을 1회만 해 두고 그 이후는 응용작업으로 처리
 - 최종 결과물은 사전형태 변수로 만들어 JSON 파일로 저장
 
기초 데이터 : 국립국어원 언어정보나눔터 전자사전 파일
 - 파일 개수 : 35개 (총 18종 = 17*2 + 1) 
 - '고유명사' 외에는 모든 파일이 기초/상세 구조로 되어 있음
 - 전체 목록 : 고유명사, (가나다순) 감탄사, 관용표현, 관형사, 대명사, 복합명사구, 부사, 분류사, 수사, 어근, 어미, 연어, 용언, 의존명사, 접사, 조사, 체언, 특수어
 
분류(확정적인 것이 아님!)

```python
import json
from collections import OrderedDict

전자사전경로 = './step1_data/'
전자사전결과 = OrderedDict()
```

```python
def 리스트반환(파일명, 기존리스트):
    if not 파일명:
        raise "오류 : 파일명은 필수 항목입니다"

    리스트 = []

    if 기존리스트:
        print('\nLog : ' + '기존 리스트 입력됨')
        리스트 = 기존리스트
    
    with open(파일명) as NLP_FILE:
        for line in NLP_FILE:
            줄바꿈제거 = line.replace('\n', '')

            if 줄바꿈제거 == '':
                continue

            if 줄바꿈제거 not in 리스트:
                if 기존리스트: # 금차 읽은 파일의 내용이 기존리스트에 없던 내용이면 출력
                    print('Log : ' +'기존 리스트에 없던 내용 : ', 줄바꿈제거)

                리스트.append(줄바꿈제거)
                
        print('Log : ' + '전체 건수 : ', len(리스트))
        return 리스트
```
